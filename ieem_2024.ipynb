```python
import numpy as np
import pandas as pd
import networkx as nx
from lingam import DirectLiNGAM
import matplotlib.pyplot as plt

# Load the dataset
selected_columns = ['ep (KW)', 'm (kg)', 'ec (cm3)', 'Ft', 'Fuel consumption ', 'Enedc (g/km)']
df = pd.read_csv("auto_co2_eur_21_raw.csv")[selected_columns]

# Handle non-numeric values
non_numeric_columns = df.select_dtypes(exclude=np.number).columns
for col in non_numeric_columns:
    if col != 'Ft':  # Assuming 'Ft' column contains the 'PETROL' values
        df = pd.get_dummies(df, columns=[col])  # One-hot encode categorical variables
    else:
        # Handle 'Ft' column separately (e.g., using label encoding)
        # Replace 'LPG' with numeric values
        df['Ft'] = df['Ft'].replace({'PETROL': 1, 'DIESEL': 0, 'PETROL/ELECTRIC': 4, 'DIESEL/ELECTRIC': 5, 'NG': 6, 'NG-BIOMETHANE': 7, 'LPG': 2, 'ELECTRIC': 3})  # Example encoding for categorical variable

# Remove rows with NaN values
df = df.dropna()

# Remove rows with 'UNKNOWN' values
df = df[~df.isin(['UNKNOWN']).any(axis=1)]

# Sample 1500 random rows
df_sample = df.sample(n=1500, random_state=42)

# Fit DirectLiNGAM model
model = DirectLiNGAM()
model.fit(df_sample)

# Generate labels for the graph
labels = df_sample.columns

# Create a directed graph
G = nx.DiGraph()

# Add nodes to the graph
G.add_nodes_from(range(len(labels)))

# Add directed edges to the graph based on the adjacency matrix
for i in range(len(labels)):
    for j in range(len(labels)):
        if model.adjacency_matrix_[i][j] != 0:
            G.add_edge(j, i, weight=model.adjacency_matrix_[i][j])  # Reverse the direction for better visualization

# Plot the causal graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  # Set seed for reproducibility
node_labels = {i: labels[i] for i in range(len(labels))}  # Create a dictionary of node labels
nx.draw(G, pos, with_labels=True, labels=node_labels, node_size=2000, node_color="skyblue", font_size=10, arrowsize=20)

# Annotate the edges with the magnitude of causality
labels = nx.get_edge_attributes(G, 'weight')
for (u, v), weight in labels.items():
    # Calculate the position for the annotation
    x = (pos[u][0] + pos[v][0]) / 2
    y = (pos[u][1] + pos[v][1]) / 2
    plt.text(x, y, f"{weight:.2f}", horizontalalignment='center', verticalalignment='center')

plt.title("Causal Graph for Selected Columns")
plt.show()

```

```python
from dowhy import CausalModel
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

def scatter_plot_with_correlation_line(x, y):
    plt.scatter(x, y, c="#c23424")
    axes = plt.gca()
    m, b = np.polyfit(x, y, 1)
    X_plot = np.linspace(axes.get_xlim()[0], axes.get_xlim()[1], 100)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.plot(X_plot, m * X_plot + b, "-")
    plt.show()

df = pd.read_csv("auto_co2_eur_21_raw2.csv", nrows=1000, usecols=['ep (KW)', 'm (kg)', 'ec (cm3)', 'Ft', 'Fuel consumption ', 'Enedc (g/km)'])
print(df.columns)

df = df.dropna()
print(df)

scatter_plot_with_correlation_line(df["m (kg)"], df['ec (cm3)'])

# DoWhy: model creation
graph_1 = """
graph [directed 1 
node [id "ep (KW)" label "ep (KW)"]
node [id "m (kg)" label "m (kg)"]
node [id "ec (cm3)" label "ec (cm3)"]
node [id "Ft" label "Ft"]
node [id "Fuel consumption " label "Fuel consumption "]
node [id "Enedc (g/km)" label "Enedc (g/km)"]
edge [source "ep (KW)" target "Fuel consumption "]
edge [source "m (kg)" target "Fuel consumption "]
edge [source "ec (cm3)" target "Fuel consumption "]
edge [source "Ft" target "Fuel consumption "]
edge [source "Enedc (g/km)" target "Fuel consumption "]
edge [source "ep (KW)" target "ec (cm3)"]
edge [source "m (kg)" target "ec (cm3)"]
edge [source "Ft" target "ec (cm3)"]
edge [source "Enedc (g/km)" target "ec (cm3)"]
]
"""

model = CausalModel(data=df, treatment="ec (cm3)", outcome="ec (cm3)", graph=graph_1)
#model.view_model()

estimand = model.identify_effect()
print(estimand)

estimate = model.estimate_effect(identified_estimand=estimand, method_name="backdoor.linear_regression")
print(estimate)

res_random = model.refute_estimate(estimand, estimate, method_name="random_common_cause")
print(res_random)

#model.view_model()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler

# Load the dataset
df = pd.read_csv("auto_co2_eur_21_raw.csv")

# Select columns for scaling
selected_columns = ['ep (KW)', 'm (kg)', 'ec (cm3)', 'Ft', 'Fuel consumption ', 'Enedc (g/km)']

# Extract the selected columns
df_selected = df[selected_columns]

# Handle non-numeric values (assuming 'Ft' column contains categorical values)
df_encoded = pd.get_dummies(df_selected, columns=['Ft'], drop_first=True)

# Sample 15000 random rows
df_sample = df_encoded.sample(n=15000, random_state=42)

# Initialize the RobustScaler
robust_scaler = RobustScaler()

# Fit and transform the selected columns
scaled_data = robust_scaler.fit_transform(df_sample)

# Convert the scaled data into a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df_sample.columns)

# Plot the scaled data using lines
plt.figure(figsize=(10, 6))
for column in scaled_df.columns:
    plt.plot(scaled_df[column], label=column)
plt.xlabel("Index")
plt.ylabel("Scaled Value")
plt.title("Scaled Data using Robust Scaler")
plt.legend()
plt.grid(True)
plt.show()
```
